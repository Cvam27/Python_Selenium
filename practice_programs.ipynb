```python
name = "shivam 100"

result = ""
for ch in name:
    if ch.isdigit():
        result += ch

print(result)
```

```python
a = 8
l1 = []
for i in range(1,a):
    if a%i==0:
        l1.append(i)
print(l1)
if sum(l1) == a:
    print("Perfect Number")
else:
    print("Not perfect")
```
